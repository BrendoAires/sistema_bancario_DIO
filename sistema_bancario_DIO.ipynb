#Projeto Bancário com Python
####Saldo, Extrato, Saque

#####-A v1 possui somente um usuário
#####-Todos os depósitos devem ser armazenados em uma variável e exibidos na operação de extrato
#####-Permitir 3 saques diários
#####-Limite máximo R$500,00 por saque
#####-Se saldo < saldo em conta, exibir mensagem de saldo insuficiente
#####-Todos os saques devem ser armazenado em uma variável e exibidos na operação de extrato
#####-O extrato deve listar todos os depósitos e saques
#####-No fim da listagem deve exibir o saldo atual

```python
menu = """
  [d] Depositar
  [s] Sacar
  [e] Extrato
  [q] Sair

  => """


extrato = ""
limite = 500.00
limite_saques = 3
valor_saque = 0.0
saldo = 0
numero_saques = 0


def depositar():
  global saldo
  global extrato

  valor_deposito = float(input('Valor a depositar: '))
  if valor_deposito > 0:
    saldo += valor_deposito
    extrato += f'Depósito de R${valor_deposito}\n'

  else:

    print("*****------Depósito não efetuado. Valor inválido*****------\n")
    valor_deposito = 0



def sacar():
  global saldo
  global extrato
  global numero_saques


  valor_saque = float(input('Digite o valor para sacar: '))

  if (valor_saque <= saldo) and valor_saque > 0:
    if (numero_saques < limite_saques):
      if (valor_saque <= limite):
        saldo = saldo - valor_saque
        extrato += f'Saque de R${valor_saque}\n'


        print("*****------Saque efetuado com sucesso------*****\n")
        numero_saques = numero_saques +1
      else:
        print("*****------Saque não efetuado. Valor ultrapassou o limite por saque------*****")
    else:
      print("*****------Você atingiu o limite de saques diários------*****")
  else:
    print('*****------Saldo Insuficiente------*****')




def ver_extrato():
  global saldo
  global extrato
  global valor_deposito
  global valor_saque




  print('Sem movimentação no periodo' if not extrato else extrato)
  print(f'**********======SALDO: R${saldo}======**********')



while True:
  opcao = input(menu)

  if opcao == "d":
    depositar()

  elif opcao == "s":
    sacar()

  elif opcao == "e":
    ver_extrato()

  elif opcao == "q":
    break

  else:
    print("Opção Inválida\n")

```
